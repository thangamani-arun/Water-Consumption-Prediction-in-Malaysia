```python
# Before any further analysis, we check for missing values across datasets

# Check for missing values in each dataset
print(population_malaysia.isnull().sum())
print(population_state.isnull().sum())
print(water_production.isnull().sum())
print(water_access.isnull().sum())
print(water_consumption.isnull().sum())
```

The output shows that there are no missing values in any of the columns in our datasets.

```python
# Convert 'date' columns to datetime format for accurate data analysis
population_malaysia['date'] = pd.to_datetime(population_malaysia['date'])
population_state['date'] = pd.to_datetime(population_state['date'])
water_access['date'] = pd.to_datetime(water_access['date'])
water_consumption['date'] = pd.to_datetime(water_consumption['date'])
water_production['date'] = pd.to_datetime(water_production['date'])
```

Filter the unnecessary data => **Feature Selections**

```python
# Function to calculate metrics
import numpy as np
def evaluate_model(y_true, y_prediction):
    mae = mean_absolute_error(y_true, y_prediction)
    mse = mean_squared_error(y_true, y_prediction)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_prediction)
    return {"MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2}

# Metrics for Linear Regression
lr_metrics = evaluate_model(y_test, y_pred_lr)

# Metrics for Random Forest
rf_metrics = evaluate_model(y_test, y_pred_rf)

# Combine results into a DataFrame
results = pd.DataFrame([lr_metrics, rf_metrics], index=["Linear Regression", "Random Forest"])
print(results)
```

### LSTM

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Normalize the data to scale the values between 0 and 1 for better LSTM performance
df_lstm = filter_model_data[['water_production', 'population', 'water_consumption' ]]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_lstm[['water_production', 'population', 'water_consumption']])

# Split data into features (X) and target (y)
X = scaled_data[:, :2]  # Features: water_production and population
y = scaled_data[:, 2]   # Target: water_consumption
```

```python
# Function to create sequences for LSTM
def create_dataset(X, y, time_step=3):
    X_seq, y_seq = [], []
    for i in range(len(X) - time_step):
        X_seq.append(X[i:i + time_step])
        y_seq.append(y[i + time_step])  # Target is the next value
    return np.array(X_seq), np.array(y_seq)

time_step = 3
X_seq, y_seq = create_dataset(X, y, time_step)

# Reshape X to be [samples, time steps, features]
X_seq = X_seq.reshape(X_seq.shape[0], X_seq.shape[1], X.shape[1])

# Split the data into training and testing sets (70% train, 30% test)
train_size = int(len(X_seq) * 0.7)
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model_lstm.add(Dropout(0.2))
model_lstm.add(LSTM(units=100, return_sequences=False))
model_lstm.add(Dense(units=1))  # Output layer for regression

# Compile the model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model_lstm.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Make predictions
y_pred_lstm = model_lstm.predict(X_test)

# Rescale predictions and actual values back to original scale
y_pred_rescaled = scaler.inverse_transform(
    np.hstack((X_test[:, -1, :], y_pred_lstm))
)[:, -1]

y_test_rescaled = scaler.inverse_transform(
    np.hstack((X_test[:, -1, :], y_test.reshape(-1, 1)))
)[:, -1]

# Evaluate the model
mse_lstm = mean_squared_error(y_test_rescaled, y_pred_rescaled)
r2_lstm = r2_score(y_test_rescaled, y_pred_rescaled)

print(f'Mean Squared Error: {mse_lstm}')
print(f'R-squared: {r2_lstm}')
```

```python
# Function to calculate metrics
import numpy as np
def evaluate_model(y_true, y_prediction):
    mae = mean_absolute_error(y_true, y_prediction)
    mse = mean_squared_error(y_true, y_prediction)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_prediction)
    return {"MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2}

# Metrics for LSTM
lstm_metrics = evaluate_model(y_test_rescaled, y_pred_rescaled)

# Metrics for ARIMA
arima_metrics = evaluate_model(test['water_consumption'], forecast)

# Combine results into a DataFrame
results = pd.DataFrame([lr_metrics, rf_metrics, lstm_metrics, arima_metrics], index=["Linear Regression", "Random Forest", "LSTM", "ARIMA"])
print(results)


```

# Result Evaluation and Interpretation

The evaluation is based on four metrics:

*   Mean Absolute Error (MAE): Average magnitude of errors, where lower values are better.
*   Mean Squared Error (MSE): Penalizes larger errors more, where lower values are better.
*   Root Mean Squared Error (RMSE): The square root of MSE, providing error magnitudes in the same unit as the dependent variable. Lower values are better.
*   R (Coefficient of Determination): Explains the variance (accuracy) in the data; values closer to 1 indicate better fit.


1. Linear Regression:
  - Exceptional predictive accuracy for water usage in Malaysia.
  - Low error metrics which indicate that the model consistently produces predictions close to the actual data points.
  - The model explains 96.1% of the variance in water consumption, highlighting its strong ability to capture the relationship between population, water production and water consumption.
  - This performance suggests that the underlying patterns in water consumption are largely linear, making Linear Regression a highly suitable choice for this dataset.

  Limitations:
  - Assumes a linear relationship between population, water production and water consumption which it may fail to capture more complex, non-linear patterns or events like sudden policy changes or global pandemic outbreaks.

  Recommendations:
  - Require regular assessment of its performance, especially if the nature of the predictors or water consumption patterns evolves over time.


2. Random Forest:
  - Produced relatively high error metrics, indicating poorer predictive performance compared to other models.
  - R (0.875) explains only 87.5% of the variance, which is quite high and in contrast to the value obtained from the models error metrics.
  - Its performance here suggests that it may not effectively handle the structure of this dataset.
  
  Limitations:
  - Possible overfitting, where the model focuses too closely on the training data, resulting in reduced generalizability to unseen data.
  - Overly sensitive to noise or irrelevant features in the dataset based on the high metric errors. This could be a result of lacking optimal hyperparameter tuning.

  Recommendations:
  - Require further refinement by deriving additional predictors such as seasonal indexes, lagged variables, rolling averages, or interaction terms.
  - Incorporating additional predictors, and optimizing hyperparameters such as tree depth, number of estimators, or minimum leaf size.
  - Hybrid approaches that combine Random Forest with models adept at time-series prediction, such as ARIMA or LSTM, could yield improved results.


3. LSTM (Long Short-Term Memory):
  - Produces intermediate MAE (224.084) and RMSE (242.500), with moderate MSE (58806.312) values.
  - Low R (0.627) suggests the model also struggles to explain the variance in the data.

  Limitations:
  - A data hungry model that requires large datasets to achieve high accuracy, which is not reflected here in our project.
  - Highly sensitive where the model requires meticulous tuning of hyperparameters, while improper configurations can lead to suboptimal results.

  Recommendations:
  - Requires optimal implementation of larger datasets, refined preprocessing, and careful tuning for LSTM could outperform other models, especially for datasets with complex temporal patterns and dependencies.

4. ARIMA (AutoRegressive Integrated Moving Average):
  - Demonstrated strong performance with the lowest MAE (170.251) and RMSE (198.809), and relatively low MSE (39,524.914).
  - It indicates that the model is suitable for univariate time-series data which highlights ARIMA's effectiveness in capturing trends and seasonality in univariate time-series data.
  - However, its moderate R value (0.748) indicates that it explains 74.8% of the variance, which is lower than Linear Regression and Random Forest.

  Limitations:
  - Inability to handle external predictors using population and water consumption data unlike other models.
  - Struggles with complex, non-linear relationships or external predictors.
  - Less versatile compared to other models.

  Recommendations:
  - Hybrid approaches that combine other machine learning models such as SARIMA or ARIMAX could yield improved results.
